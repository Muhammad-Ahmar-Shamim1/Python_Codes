```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
x =[2,3,4,6]
y =[10,20,30,40]

plt.plot(x,y,color ="purple")
plt.title("Simple Line Plot")
plt.xlabel("X -axies")
plt.ylabel("Y -axies")
plt.show()
```

```python
plt.figure(figsize=(8,5))
x =[2,3,4,6]
y =[10,20,30,40]

plt.bar(x,y,color ="purple")
plt.title("Simple Line Plot")
plt.xlabel("X -axies")
plt.ylabel("Y -axies")
plt.show()
```

```python
plt.figure(figsize=(8,5))
x =[2,3,4,6]
y =[10,20,30,40]

plt.barh(x,y,color ="purple")
plt.title("Simple Line Plot")
plt.xlabel("X -axies")
plt.ylabel("Y -axies")
plt.show()
```

```python
labels =['Apple','Banana','Cherries','Dates']
sizes =[23,40,35,20]
plt.pie(sizes ,labels = labels,autopct='%1.1f%%',startangle=(90),explode =[0,0.1,0,0])
plt.title("Fruit Distribution")
plt.show()
```
